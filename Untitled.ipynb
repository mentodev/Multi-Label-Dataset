```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import train_test_split
import cv2
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("New_Train.csv")
```

```python
train
```

```python
train_image = []
for i in tqdm(range(train.shape[0])):
    img = tf.keras.preprocessing.image.load_img('Images_new/'+train['Id'][i]+'.jpg',target_size=(224,224,3))
    img = tf.keras.preprocessing.image.img_to_array(img)
    img = img/255
    train_image.append(img)
X = np.array(train_image)
```

```python
X_new = X[:100]
X_new.shape
```

```python
plt.imshow(X[0])
```

```python
train['Genre'][2]
```

```python
y = np.array(train.drop(['Id', 'Genre'],axis=1))
y_new = y[:100]
y_new.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, random_state=42, test_size=0.1)
```

```python
model = tf.keras.Sequential()
```

```python
input_shape = None
type(input_shape)
```

```python
def create_Conv_Layers(filters, kernel_size, pool_size, activation, input_shape, i):
    if type(input_shape) == tuple:
        model.add(tf.keras.layers.Conv2D(filters=filters,kernel_size = kernel_size,activation = activation,input_shape = input_shape))
    else:
        model.add(tf.keras.layers.Conv2D(filters = filters,kernel_size=kernel_size,activation = activation))
    model.add(tf.keras.layers.MaxPool2D(pool_size=pool_size, padding = 'same'))
    if i%2 == 0:
        model.add(tf.keras.layers.Dropout(0.2))
        
def create_Dense_layers(units, activation):
    model.add(tf.keras.layers.Dense(units = units, 
                                   activation = activation))
```

```python
create_Conv_Layers(filters=16, kernel_size=(3, 3), pool_size=(2, 2), activation='relu', input_shape=(224, 224, 3), i=1)
for i in range(2, 4, 1):
    create_Conv_Layers(filters = 64, kernel_size=(3, 3), pool_size=(2, 2), activation='relu', input_shape=1, i=i)

model.add(tf.keras.layers.Flatten())    
for i in range(1, 4, 1):
    create_Dense_layers(units = np.power(2, i)*32, activation='relu')
create_Dense_layers(units = 25, activation='sigmoid')
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 10, validation_data = (X_test, y_test))
```
